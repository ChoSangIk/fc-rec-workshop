```python
import random as rd
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('movies_w_imgurl.csv')
```

```python
movies['genreCount'] = movies['genres'].apply(lambda x: len(x.split('|')))
```

```python
chosenItems = []
for row in movies[movies['genreCount'] > 1]['movieId']:
    if len(chosenItems) >= 5:
        break;
    if rd.random() <= 0.1:
        chosenItems.append(row)
chosenItems
```

```python
from IPython.display import Image, display
movies[movies['movieId'].isin(chosenItems)]
for idx in movies[movies['movieId'].isin(chosenItems)].index:
    item = movies.iloc[idx,:]
    display(Image(item.imgurl))
    print(item)
```
