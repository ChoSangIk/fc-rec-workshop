```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ratings = pd.read_csv('ratings-9_1.csv')
```

```python
trainRatingCounts = pd.DataFrame(ratings[ratings['type'] == 'train'].groupby('userId')['rating'].count())
```

```python
chosenUser = []
for user in trainRatingCounts[trainRatingCounts['rating'] > 20].index:
    if len(chosenUser) >= 5:
        break;
    if random.random() <= 0.02:
        chosenUser.append(user)
chosenUser
```
